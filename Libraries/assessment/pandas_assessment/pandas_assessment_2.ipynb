```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rs=pd.read_excel("/content/retail_data.xlsx")
rs
```

```python
# 1. View the structure of the dataset (columns, types, missing values)
display(rs.columns)
display(rs.dtypes)
display(rs.isnull().sum())
```

```python
# 2. What is the shape (rows, columns) of the dataset?
display(rs.shape)
display(rs.columns)
```

```python
# 3. Are there any duplicate records?
display(rs.duplicated().sum())
```

```python
# 4.  Are there any missing or corrupted entries in Ship Date, Order Date, or numeric columns?
display(rs.isnull().sum())
```

```python
# 5.  Convert Order Date and Ship Date to datetime.
rs['Order Date'] = pd.to_datetime(rs['Order Date'])
rs['Ship Date'] = pd.to_datetime(rs['Ship Date'])
rs
```

```python
# 6.  Check for future or inconsistent shipping dates.
display(rs[rs['Ship Date'] < rs['Order Date']])
```

```python
# 7. Convert price columns to numeric (remove $ and commas).
The columns 'Cost Price' and 'Retail Price' are already in a numeric format, so no conversion is needed.
rs['Cost Price'] = rs['Cost Price'].str.replace('$', '').str.replace(',', '').astype(float)
rs['Retail Price'] = rs['Retail Price'].str.replace(',', '').astype(float)
rs
```

```python
# 8. What are the unique values in Customer Type and Order Priority?
display(rs['Customer Type'].unique())
display(rs['Order Priority'].unique())
```

```python
# 9. What are the most common shipping modes?
display(rs['Ship Mode'].value_counts())
```

```python
# 10. Which cities have the highest number of orders?
display(rs['City'].value_counts())
```

```python
# 11. Whatâ€™s the range of order quantities and prices?
display(rs['Order Quantity'].describe())
display(rs['Cost Price'].describe())
display(rs['Retail Price'].describe())
```

```python
# 12. Create a new column for shipping duration
rs['Shipping Duration'] = rs['Ship Date'] - rs['Order Date']
rs
```

```python
# 13. Are there any orders with zero or negative total or quantity?
display(rs[rs['Total'] <= 0])
display(rs[rs['Order Quantity'] <= 0])
```

```python
# 14. Are all discount percentages matching discount dollar amounts?
display(rs[rs['Discount %'] != rs['Discount $']])
```

```python
# 15.  Check for mismatches in total calculation
rs['Total_calculated'] = rs['Order Quantity'] * rs['Retail Price']
display(rs[rs['Total'] != rs['Total_calculated']])
```

```python
# 16.  Identify top 5 products by order quantity.
display(rs.groupby('Product Name')['Order Quantity'].sum().sort_values(ascending=False).head(5))
```

```python
# 17. Which Account Manager handled the most revenue?
display(rs.groupby('Account Manager')['Total'].sum().sort_values(ascending=False))
```

```python
# 18.  What is the average shipping cost by mode?
display(rs.groupby('Ship Mode')['Shipping Duration'].mean())
```

```python
# 19. Find the most profitable product.
display(rs.groupby('Product Name')['Total'].sum().sort_values(ascending=False))
```

```python
# 20. Find the most profitable product.
display(rs.groupby('Product Name')['Total'].sum().sort_values(ascending=False))
```

```python
# 1. What is the total revenue generated across all order?
display(rs['Total'].sum())
```

```python
# 2. Which customer type generates more revenue?
display(rs.groupby('Customer Type')['Total'].sum().sort_values(ascending=False))
```

```python
# 3.  How does order priority affect revenue?
display(rs.groupby('Order Priority')['Total'].sum().sort_values(ascending=False))
```

```python
# 4. What is the average profit margin by product category?
rs['Profit Margin'] = (rs['Retail Price'] - rs['Cost Price']) / rs['Retail Price']
display(rs.groupby('Product Category')['Profit Margin'].mean())
```

```python
# 5. What is the most profitable product overall?
display(rs.groupby('Product Name')['Total'].sum().sort_values(ascending=False))
```

```python
# 6. How many days does it usually take to ship an order?
display(rs['Shipping Duration'].mean())
```

```python
# 7. Do longer shipping times impact profit margins?
display(rs.groupby('Shipping Duration')['Profit Margin'].mean())
```

```python
#  8. Which city brings in the highest revenue?
display(rs.groupby('City')['Total'].sum().sort_values(ascending=False))
```

```python
# 9. Which account manager generated the most revenue?
display(rs.groupby('Account Manager')['Total'].sum().sort_values(ascending=False))
```

```python
# 10.  Which shipping mode is most cost-effective (lowest avg. shipping)?
display(rs.groupby('Ship Mode')['Shipping Duration'].mean())
```

```python
# 11.  Do higher discounts reduce profits?
display(rs.groupby('Discount %')['Profit Margin'].mean())
```

```python
# 12. Which state has the highest number of orders?
display(rs['State'].value_counts())
```

```python
# 13. What is the average discount % across all orders?
display(rs['Discount %'].mean())
```

```python
# 14. What is the average total spend per order?
display(rs['Total'].mean())
```

```python
# 15. . Are certain containers (e.g., Small Box, Wrap Bag) more profitable?
display(rs.groupby('Product Container')['Total'].sum().sort_values(ascending=False))
```
