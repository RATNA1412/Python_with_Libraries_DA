```python
# 14. Which indicator has the highest correlation with stock close price?
# Exclude non-numeric columns before calculating correlation
numeric_rs = rs.select_dtypes(include=np.number)
correlation_with_close_price = numeric_rs.corrwith(numeric_rs['Close Price'])
print(correlation_with_close_price)
```
