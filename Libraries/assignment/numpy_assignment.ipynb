```python
# 18. Describe a practical example where NumPy can be used in EDA, AI, ML, and DL, and implement a NumPy solution for a simple task in each area.
     # 1. Exploratory Data Analysis (EDA):
ages = np.array([25, 30, 22, 40, 35, 28, 60, 45])

mean_age = np.mean(ages)
median_age = np.median(ages)
std_age = np.std(ages)

print("Mean age:", mean_age)
print("Median age:", median_age)
print("Standard Deviation:", std_age)

      # 2. Artificial Intelligence (AI)
utility_scores = np.array([0.4, 0.7, 0.6, 0.8, 0.5])
best_move = np.argmax(utility_scores)
print("Best move index:", best_move)
print("Best move utility:", utility_scores[best_move])

      #  3. Machine Learning (ML)
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])
X = np.vstack([np.ones(len(x)), x]).T
w = np.linalg.inv(X.T @ X) @ X.T @ y
print("Linear regression weights:", w)

      # 4. Deep Learning (DL)
inputs = np.array([[1.0, 2.0, -1.0],
                   [0.5, 0.0, 2.0]])
weights = np.array([[0.2, -0.5],
                    [0.3, 0.8],
                    [-0.6, 0.1]])
biases = np.array([0.1, -0.2])
z = inputs @ weights + biases
output = np.maximum(0, z)
print("Output of dense layer with ReLU:")
print(output)
```

```python
# 19. Create a random 4x4 matrix and use NumPy to compute its eigenvalues and eigenvectors.
#  Verify the eigenvalues by reconstructing the matrix.

np.random.seed(42)

A = np.random.rand(4, 4)
print("Original Matrix A:\n", A)


eigenvalues, eigenvectors = np.linalg.eig(A)

print("\nEigenvalues:\n", eigenvalues)
print("\nEigenvectors (columns):\n", eigenvectors)


D = np.diag(eigenvalues)
V = eigenvectors
V_inv = np.linalg.inv(V)

A_reconstructed = V @ D @ V_inv


print("\nReconstructed Matrix A:\n", A_reconstructed)


print("\nAre A and reconstructed A close?", np.allclose(A, A_reconstructed))
```

```python
# 22. Create a 3D array of shape (2, 1, 4) and a 2D array of shape (4, 1).
# Perform an element-wise operation using broadcasting and explain the result.
# Use np. new axis to achieve the same result without broadcasting.
A = np.array([[[1, 2, 3, 4]],
              [[5, 6, 7, 8]]])

B = np.array([[10],
              [20],
              [30],
              [40]])
B_T = B.T


result_broadcast = A + B_T

print("Broadcasted result:\n", result_broadcast)
print("Shape:", result_broadcast.shape)

B_reshaped = B.T[np.newaxis, :, :]


result_manual = A + B_reshaped

print("\nManual reshape with np.newaxis:\n", result_manual)
print("Shape:", result_manual.shape)
```

```python
# 23. Generate a 2D array of random floats between 0 and 1.
# Use conditional operators to create a Boolean mask for values less than 0.5.
# Replace these values with their squares and leave the rest unchanged.

np.random.seed(42)
arr = np.random.rand(4, 5)

print("Original Array:\n", arr)

mask = arr < 0.5


arr[mask] = arr[mask] ** 2

print("\nModified Array (values < 0.5 squared):\n", arr)
```
